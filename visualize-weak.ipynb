```python
from simulationTools import *
from scipy.signal import medfilt
```

```python
tList = np.load('./weak_data/tList.npy')
LHSneu = np.load('./weak_data/LHSneu.npy')
LHSmm = np.load('./weak_data/LHSmm.npy')
f = np.load('./weak_data/f.npy')
fOrd = np.load('./weak_data/fOrd.npy')
fTerm = np.load('./weak_data/fTerm.npy')
```

```python
# Plot both sides of the inequality
ax = plt.subplot(111)
ax.plot(tList, LHSmm, label='Min-max', ls='dashdot', lw=2)
ax.plot(tList, LHSneu, label='von Neumann', ls='dashed', lw=2)
ax.plot(tList, f, label='RHS', lw=2)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.ylim(0)
plt.xlabel("Time (units of 1/J)")
plt.ylabel("Bits")
plt.legend(frameon=False)
plt.savefig('LHSs_RHS.pdf')
plt.show()
```

```python
# Plot the RHS terms dependent on g
ax = plt.subplot(111)
ax.plot(tList, fOrd[:,1], label='Terms \propto g', ls='dashed', lw=2)
ax.plot(tList, fOrd[:,2], label='Terms \propto g^{2}', ls='dashdot', lw=2)
ax.plot(tList, fOrd[:,1]+fOrd[:,2], label='Sum', lw=2)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.xlabel("Time (units of 1/J)")
plt.ylabel("Bits")
plt.legend(frameon=False)
plt.savefig('RHS_Zoom.pdf')
plt.show()
```

```python
# Plot the contributions to the RHS by
# terms which include quasiprobabilities
ax = plt.subplot(111)
ax.plot(tList, fOrd[:,2], label='All Terms \propto g^{2}', ls='solid', lw=2)
ax.plot(tList, medfilt(fTerm[:,3],3), label='Term \propto A_{1213}', ls='dashed', lw=2)
ax.plot(tList, medfilt(fTerm[:,4],3), label='Term \propto A_{2223}', ls='dotted', lw=2.5)
ax.plot(tList, medfilt(fTerm[:,5],3), label='Term \propto A_{1223}', ls='dashdot', lw=2)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.xlabel("Time (units of 1/J)")
plt.ylabel("Bits")
plt.legend(frameon=False)
plt.savefig('Quasiprobs.pdf')
plt.show()
```
